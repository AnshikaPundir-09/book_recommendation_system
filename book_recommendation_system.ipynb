# popularity based recommender system

```python
rating_data=rating.merge(books,on='ISBN')
```

```python
rating_data.head()
```

```python
number_rating=rating_data.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
number_rating.rename(columns={'Book-Rating':'rating_count'})
```

```python
#  book rating  data kuch  jgh pe numberic nahi hai .. convert it to numeric first
rating_data['Book-Rating'] = pd.to_numeric(rating_data['Book-Rating'], errors='coerce')
avg_rating = rating_data.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
avg_rating
```

```python
popular_df=number_rating.merge(avg_rating,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['Book-Rating']>=200].sort_values('avg_rating',ascending=False).head(100) 
```

```python
readers=x[x].index
```

```python
rating_data['User-ID'].isin(readers)
```

```python
rating_data[rating_data['User-ID'].isin(readers)]
```

```python
filtered_rating=rating_data[rating_data['User-ID'].isin(readers)]
```

```python
#  filtering on the basis of books
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
famous_books=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(pt)
```

```python
 similarity_score=cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recommend(book_name):
    try:
        
        index = np.where(pt.index == book_name)[0][0]
        distances = similarity_score[index]
        similar_items = sorted(list(enumerate(distances)), key=lambda x: x[1], reverse=True)[1:8]

        print(f"\nTop recommendations for '{book_name}':\n")
        for i in similar_items:
            print(pt.index[i[0]])

    except IndexError:
        print(f"Book '{book_name}' not found in the dataset.")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    book_name = input("Enter a book name: ")
    recommend(book_name)
```
